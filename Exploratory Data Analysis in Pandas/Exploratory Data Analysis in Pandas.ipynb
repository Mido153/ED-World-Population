```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.float_format' , lambda x: '%.2f' % x)
```

## Read the data 

```python
df = pd.read_csv('world_population.csv')
df
```

```python
df.describe()
```

## All unique values for each column 

```python
df.nunique()
```

## Population on 2022 sort

```python
df.sort_values(by='2022 Population' , ascending=False).head(5)
```

## Correlation between the data columns 

```python
df.corr()
```

## Plot the heat map for the correlation

```python
sns.heatmap(df.corr() , annot=True)
# plt.rcParams['figure.figsize'] = (10,7)
plt.figure(figsize=(10,7))
plt.show()
```

## Group data by continent to see the populations 

```python
df.groupby('Continent').mean().sort_values(by='2022 Population'  , ascending=False)
```

```python
df.columns
```

```python
df2 = df.groupby('Continent')['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population'].mean().sort_values(by='2022 Population'  , ascending=False)
df2 = df2.transpose()
df2
```

```python
plt.figure(figsize=(15,6))
plt.plot(df2);
```
